```python
with open('14.txt') as f:
    robots = [list(map(int,
        line.replace('p=', '')
            .replace('v=', '')
            .replace(',', ' ').split()))
        for line in f.readlines()]
```

```python
WIDTH = 101
HEIGHT = 103
def moves(robots: list[int], N: int):
    return [
        ((x+vx*N) % WIDTH, (y+vy*N) % HEIGHT)
        for x, y, vx, vy in robots]
```

```python
def display(robots: list[tuple[int, int]]):
    grid = [[0 for _ in range(WIDTH)] for _ in range(HEIGHT)]
    for x, y in robots:
        grid[y][x] += 1
    return '\n'.join(
        ''.join(
            str(d) if d else '.' for d in line
    ) for line in grid)

# print(display(moves(robots, 100)))
```

# Part 1: Safety quadrants

```python
from functools import reduce
from operator import mul
product = lambda seq: reduce(mul, seq, 1)

def quadrant(x: int, y: int):
    if x == WIDTH // 2 or y == HEIGHT // 2:
        return None
    return (x < WIDTH // 2, y < HEIGHT // 2)

quadrants = {}
for x, y in moves(robots, 100):
    if q := quadrant(x, y):
        quadrants.setdefault(q, 0)
        quadrants[q] += 1
product(quadrants.values())
```
